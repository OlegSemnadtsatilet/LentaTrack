```python
dataset = pd.read_csv("C:/Users/Полина/Downloads/pricing_hackathon-20211210T173728Z-001/pricing_hackathon/pricing_hackathon_checks_prices.csv")
dataset.head()
```

```python
dataset = dataset.drop(columns=["Unnamed: 0", "Unnamed: 0.1"], axis=1)
dataset.head()
```

```python
dataset.columns
```

```python
dataset = dataset.drop(columns=["client_id", "time", "region_name", "is_ex_billa", "is_store", "division_id", "region_id", 'reg_price', 'promo_price', 'reg_card_price' ], axis=1)
dataset.head()
```

```python
encoder = LabelEncoder()
for column in dataset:
    if column in ["day", "shop_id", "check_id", "sku", "promo_id", "city_id", "class", "group", "category", "subcategory"]:
        dataset[column] = encoder.fit_transform(dataset[column])
dataset.head()
```

```python
dataset_sample = dataset.sample(50000)
dataset_sample2 = dataset.sample(50000)
y_train = dataset_sample["num_sales"]
x_train = dataset_sample.drop(columns=["num_sales", "check_pos"], axis=1)
y_test = dataset_sample2["num_sales"]
x_test = dataset_sample2.drop(columns=["num_sales", "check_pos"], axis=1)
```

```python
estimators_list = [250]
scores = {}

for estimators in estimators_list:
    start_time = datetime.datetime.now()
    gradient_boost = GBR(max_depth=5, n_estimators=estimators, learning_rate=1.0)
    gradient_boost.fit(x_train, y_train)
    y_predict = gradient_boost.predict(x_test)
    MSE = mean_squared_error(y_predict, y_test)
    scores[estimators] = MSE
    
    print("R2 :", r2_score(y_test, y_predict))
    print("MSE n =", estimators, MSE)
    print(f"Time elapsed: {datetime.datetime.now() - start_time}")
```

```python
def wape(y_true: np.array, y_pred: np.array):
    return np.sum(np.abs(y_true - y_pred)) / np.sum(np.abs(y_true))
print(wape(y_test, y_predict))
```
